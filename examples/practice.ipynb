```python
def sortedSquares(nums):
    l = len(nums)
    
    i = 0
    while nums[i] < 0 and i < l:
        print("hola", i, nums[i])
        i+=1
        
    j = i-1
    print(i, j)

    res = []
    while i<l and j >= 0:
        if abs(nums[i]) < abs(nums[j]):
            res.append(nums[i]**2)
            i+=1
        elif abs(nums[i]) > abs(nums[j]):
            res.append(nums[j]**2)
            j-=1
        else:
            res.append(nums[i]**2)
            res.append(nums[j]**2)
            i+=1
            j-=1
            
    while i < l:
        res.append(nums[i]**2)
        i+=1
        
    while j >= 0:
        res.append(nums[j]**2)
        j-=1
        
    return res
```

```python
def getAverages(nums, k: int):
    length = len(nums)
    
    runningSum = [nums[0]]
    for num in nums[1:]:
        runningSum.append(runningSum[-1] + num)
    
    print("nums: ", nums)
    print("K: ", k)
    print("runningSum: ", runningSum)
    
    mov_avg = []
    for i in range(0,length):
        print("\n",i)
        if i-k<0:
            mov_avg.append(-1)
        elif i+k>=length:
            mov_avg.append(-1)
        else:
            if i-k==0:
                total = runningSum[i+k]
            else:
                total = runningSum[i+k] - runningSum[i-k-1]
            mov_avg.append(int(total/(2*k+1)))
            print("total: ", total)
            
    return mov_avg

getAverages([7,4,3,9,1,8,5,2,6], 3)
```

```python
from collections import defaultdict 

def findWinners(matches):
        players = set()
        loss_count = defaultdict(int)
        
        for match in matches:
            players.add(match[0])
            players.add(match[1])
            loss_count[match[1]]+=1
            
        # sort the loss_count dictionary by value
        loss_count = dict(sorted(loss_count.items()))
        print(loss_count)
        
        no_loss, one_loss = players, []
        for loser, losses in loss_count.items():
            if losses > 0:
                no_loss.remove(loser)
                if losses == 1:
                    one_loss.append(loser)
                
        no_loss = list(no_loss)
        
        return [no_loss, one_loss]

findWinners([[1,3],[2,3],[3,6],[5,6],[5,7],[4,5],[4,8],[4,9],[10,4],[10,9]])
```

```python
from collections import defaultdict 

def lengthOfLongestSubstring(s: str) -> int:
        dic = defaultdict(int)
        i = 0
        j = 0
        max_len = 0
        curr_len = 0
        while j < len(s):
            print(i,j)
            curr_len+=1
            if s[j] in dic:
                while dic[s[j]] > 0:
                    dic[s[i]]-=1
                    curr_len-=1
                    i+=1
            else:
                dic[s[j]] = 1

            max_len = max(max_len, curr_len)
            j+=1
 
        return max_len

s = "abcabcbb"
# s = "dvdf"
lengthOfLongestSubstring(s)
```
